# Random Walk 

In mathematics, a random walk is a mathematical object, known as a stochastic or random process, that describes a path that consists of a succession of random steps on some mathematical space such as the integers.

An elementary example of a random walk is the random walk on the integer number line, which starts at 0 and at each step moves +1 or âˆ’1 with equal probability. Other examples include the path traced by a molecule as it travels in a liquid or a gas (see Brownian motion), the search path of a foraging animal, the price of a fluctuating stock and the financial status of a gambler: all can be approximated by random walk models, even though they may not be truly random in reality.

# 1D random walk

```python
# Libraries
import numpy as np 
import matplotlib.pyplot as plt 
```

### Create the initial position and random steps

```python
# Starting position  
position = [0] 

# Array of random steps between 0 an 1 
steps = np.random.random(500)
```

### Probability to move up or down
* Suppose to add up to 1 for unbiased random waking.
* The absolute difference between the two probabilities for moving up and down is the probablity to stay in the same place.

```python
step_up = steps > 0.9
step_down = steps < 0.1 
```

### 1D Random walk between two limits

```python
# Space limits
Max = 1000
Min = -100

for up, down in zip(step_up, step_down): 
    
    # Only go up and down if you are in the limit
    move_up = up and position[-1] < Max
    move_down = down and position[-1] > Min
    
    # Append the next position
    position.append(position[-1] + move_up - move_down) 
```

### 1D random walk graph

```python
plt.figure(figsize=(10,6))

plt.plot(position)
plt.title('1D random walk')
plt.ylabel('Space')
plt.xlabel('Time')
plt.grid(axis='y')
plt.show()
```

# 2D random walk

### Create the initial position and random steps

```python
# Starting position  
position = [[0, 0]] 

# Array of random steps between 0 an 10 
steps = np.random.randint(0,10,1000000)
```

### Probability to move up, down, right or left 
* One value for each direction for unbiased random waking.
* The rest of the values are the probability to stay.

```python
# Bias the random walk to move more horizontaly

step_up = steps == 2          # move up at value 2
step_down = steps == 7        # move down at value 7
step_right = steps < 2        # move right at value 0, 1
step_left = steps > 7         # move left at value 8, 9
```

### 2D Random walk on limited plan

```python
# plan limits
Max_horizonal = 2000
Min_horizonal = -200
Max_vertical = 1000
Min_vertical = -100

for up, down, right, left in zip(step_up, step_down, step_right, step_left): 
    
    # Move only in the limit
    move_right = right and position[-1][0] < Max_horizonal
    move_left = left and position[-1][0] > Min_horizonal
    move_up = up and position[-1][1] < Max_vertical
    move_down = down and position[-1][1] > Min_vertical
    
    # Append the next position
    position.append([position[-1][0] + move_right - move_left, position[-1][1] + move_up - move_down])
    

# Transform the position list into numpy array
position = np.array(position)
```

### Draw the 2D random walk path on the plan

```python
plt.figure(figsize=(10,6))

plt.scatter(position[:,0],position[:,1])
plt.title('2D random walk')
plt.ylabel('Y')
plt.xlabel('X')
plt.grid()
plt.show()
```
